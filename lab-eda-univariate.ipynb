```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis

# Load the dataset
df = pd.read_csv('C:\Users\Luigi\Documents\GitHub\IronHack (Data-Bootcamp-v4_April-June-2024)\Week 5\1.lab-eda-univariate.csv')

# Part 1: Understanding Product Categories
# Frequency table for product categories
category_counts = df['category'].value_counts()

# Top 5 most listed product categories
top_5_categories = category_counts.head(5)
print(top_5_categories)

# Bar chart for product categories
plt.figure(figsize=(12, 8))
sns.barplot(x=category_counts.index, y=category_counts.values)
plt.title('Distribution of Products Across Categories')
plt.xlabel('Category')
plt.ylabel('Number of Listings')
plt.xticks(rotation=90)
plt.show()

# Pie chart for top 5 categories
plt.figure(figsize=(8, 8))
top_5_categories.plot.pie(autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Top 5 Product Categories')
plt.ylabel('')
plt.show()

# Part 2: Delving into Product Pricing
# Measures of centrality for product prices
mean_price = df['price'].mean()
median_price = df['price'].median()
mode_price = df['price'].mode()[0]

print(f"Mean Price: {mean_price}")
print(f"Median Price: {median_price}")
print(f"Mode Price: {mode_price}")

# Measures of dispersion for product prices
variance_price = df['price'].var()
std_dev_price = df['price'].std()
range_price = df['price'].max() - df['price'].min()
iqr_price = df['price'].quantile(0.75) - df['price'].quantile(0.25)

print(f"Variance: {variance_price}")
print(f"Standard Deviation: {std_dev_price}")
print(f"Range: {range_price}")
print(f"IQR: {iqr_price}")

# Histogram for product prices
plt.figure(figsize=(12, 8))
sns.histplot(df['price'], bins=50, kde=True)
plt.title('Distribution of Product Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

# Box plot for product prices
plt.figure(figsize=(12, 8))
sns.boxplot(x=df['price'])
plt.title('Box Plot of Product Prices')
plt.xlabel('Price')
plt.show()

# Part 3: Unpacking Product Ratings
# Measures of centrality for product ratings
mean_rating = df['rating'].mean()
median_rating = df['rating'].median()
mode_rating = df['rating'].mode()[0]

print(f"Mean Rating: {mean_rating}")
print(f"Median Rating: {median_rating}")
print(f"Mode Rating: {mode_rating}")

# Measures of dispersion for product ratings
variance_rating = df['rating'].var()
std_dev_rating = df['rating'].std()
iqr_rating = df['rating'].quantile(0.75) - df['rating'].quantile(0.25)

print(f"Variance: {variance_rating}")
print(f"Standard Deviation: {std_dev_rating}")
print(f"IQR: {iqr_rating}")

# Skewness and kurtosis for product ratings
skewness_rating = skew(df['rating'].dropna())
kurtosis_rating = kurtosis(df['rating'].dropna())

print(f"Skewness: {skewness_rating}")
print(f"Kurtosis: {kurtosis_rating}")

# Histogram for product ratings
plt.figure(figsize=(12, 8))
sns.histplot(df['rating'], bins=10, kde=True)
plt.title('Distribution of Product Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```
